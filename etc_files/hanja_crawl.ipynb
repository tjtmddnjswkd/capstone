```python
import pandas as pd
import pickle
import os
import re

from tqdm.notebook import tqdm
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
```

```python
gdrive_path = r'G:\내 드라이브\KB인턴관련'
if os.path.exists(gdrive_path): pass
else: gdrive_path = "content/drive/MyDrive/KB인턴관련"

with open(os.path.join(gdrive_path, "etc_files/hanja_list.pickle"), "rb") as f:
    hz_list = pickle.load(f)
```

```python
hz_list
```

```python
driver = webdriver.Chrome()
driver.get("https://dic.daum.net/index.do?dic=hanja")

chz_dict=dict()
for hz in tqdm(hz_list):
    chz_dict[hz] = hz
    # main 페이지
    driver.get("https://dic.daum.net/index.do?dic=hanja")
    # 검색
    try:
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="q"]')))
        driver.find_element(By.XPATH, '//*[@id="q"]').send_keys(hz+"\n")
        # 검색 결과 찾기
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.NAME, 'searchWords')))
        word_search = driver.find_element(By.NAME, "searchWords")
        
        for t in word_search.find_elements(By.CLASS_NAME, "search_word"):
            w_list = t.text.split("\n")
            w_list[0] = re.sub("[0-9]+", "", w_list[0])
            chz_dict[hz] = chz_dict[hz].replace(w_list[0], w_list[1])
            if re.sub(r"[ㄱ-ㅎ|가-힣]+", "", chz_dict[hz]) == "":
                # print(f"'{hz}': '{chz_dict[hz]}'")
                break
    except:
        pass
```

```python
chz_dict
```

```python
pattern = r"[一-鿕|㐀-䶵|豈-龎]"
find_hz = re.compile(pattern)

for k in tqdm(chz_dict.keys()):
    v = chz_dict[k]
    v_list = find_hz.findall(v)
    for h in v_list:
        driver.get("https://dic.daum.net/index.do?dic=hanja")
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="q"]')))
        driver.find_element(By.XPATH, '//*[@id="q"]').send_keys(h+"\n")
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'sub_read')))
        ch = driver.find_element(By.CLASS_NAME, "sub_read").text.split(" ")[1].strip(",")
        v = v.replace(h, ch)
    chz_dict[k] = v
```

```python
for key in chz_dict.keys():
    # chz_dict[key] = chz_dict[key].replace(" ", "")
    chz_dict[key] = chz_dict[key].replace("-", "")
chz_dict
```

```python
from hangul_utils import split_syllables, join_jamos

for h in chz_dict.keys():
    v = chz_dict[h]
    jamo = split_syllables(v)
    first_chr = split_syllables(v[0])
    if jamo[0]=="ㄹ":
        if len(first_chr)==2:
            jamo = "ㅇ"+jamo[1:]
        else:
            jamo = "ㄴ"+jamo[1:]
        
        chars = list(set(jamo))
        char_to_ix = { ch:i for i,ch in enumerate(chars) }
        ix_to_char = { i:ch for i,ch in enumerate(chars) }

        jamo_numbers = [char_to_ix[x] for x in jamo]

        restored_jamo = ''.join([ix_to_char[x] for x in jamo_numbers])
        restored_text = join_jamos(restored_jamo)
        
        chz_dict[h] = restored_text
    else:
        pass
```
