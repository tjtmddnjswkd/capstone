# 브라켓 내용 제거

**특수문자 케이스**

1. <, > 브라켓 : 삭제
2. [, ] 브라켓 : 삭제
3. 고유명사 브라켓
4. dot: "로 대치

```python
import re

pattern = r'\<[^>]*\>|\[[^]]*\]'
samp = '안녕 나는 <벼벼벼볍>이야 [잘부탁해]'
re.sub(pattern=pattern, repl='', string=samp)
```

```python
df_dropna['ENG'] = df_dropna['ENG'].apply(lambda x: re.sub(pattern=pattern, repl='', string=x))
df_dropna['KOR'] = df_dropna['KOR'].apply(lambda x: re.sub(pattern=pattern, repl='', string=x))
```

```python
df_dropna["LEN_ENG"] = df_dropna["ENG"].apply(lambda x: len(x.split('\n')))
df_dropna["LEN_KOR"] = df_dropna["KOR"].apply(lambda x: len(x.split('\n')))
df_dropna
```

```python
df_dropna.loc[(df_dropna["LEN_ENG"] != df_dropna["LEN_KOR"]),:]
```

# 한 줄씩 잘라보기

```python
df_dropna.loc[123, "ENG"]
```

```python
from tqdm.notebook import tqdm

df_detail = pd.DataFrame(columns=['category', 'lawName', 'ENG', 'KOR'])

not_matched = []
not_matched_idx = 1
for idx in tqdm(df_dropna.index):
  eng_str = df_dropna.loc[idx, 'ENG'].split('\n')
  kor_str = df_dropna.loc[idx, 'KOR'].split('\n')
  cate_name, law_name = df_dropna.loc[idx, "category"], df_dropna.loc[idx, "lawName"]
  eng_len, kor_len = len(eng_str), len(kor_str)
  if eng_len!=kor_len:
    not_matched.append(idx)
    print(f"Not Matched !\n\t{not_matched_idx} - {law_name}")
    not_matched_idx+=1
    continue
  else:
    df_detail = pd.concat([df_detail, pd.DataFrame({'category':[cate_name]*eng_len, 'lawName':[law_name]*eng_len, 'ENG':eng_str, 'KOR':kor_str})], axis=0)
```

```python
df_detail.reset_index(drop=True, inplace=True)
```

```python
df_detail.tail()
```

```python
import pickle

with open(gdrive_path+"/not_matched.pkl","wb") as f:
    pickle.dump(not_matched, f)
```

# 정규식 활용

```python
check_list = ["CHAPTER", "Article", "ADDENDA", "ADDENDUM", "Section"]
```

```python
test_sent = "CHAPTER I GENERAL PROVISIONS"
test_ss = "ME ME ME"
comp = re.compile("^CHAPTER")
result = comp.match(test_ss)
print(result)
re.sub(pattern="CHAPTER", repl='', string=test_sent)
```

```python
drop_index = []
for idx in tqdm(df_detail.index):
  eng = df_detail.loc[idx, "ENG"]
  for check_word in check_list:
    comp = re.compile(f'^{check_word}')
    result = comp.match(eng)
    if result is not None:
      drop_index.append(idx)
```

```python
df_new = df_detail.loc[~df_detail.index.isin(drop_index)].reset_index(drop=True)
```
