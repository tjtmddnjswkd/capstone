```python
def count_tag(tokens):
    token_dict = dict()
    for token in tokens:
        try:
            token_dict[token.tag].append(token.form)
        except:
            token_dict[token.tag] = [token.form]
    return token_dict
```

```python
count_tokens = kiwi_tokens.apply(count_tag)
count_tokens
```

```python
nng_dict = dict()
for tokens in count_tokens.apply(lambda x: x["NNG"]):
    for t in tokens:
        try:
            nng_dict[t] += 1
        except:
            nng_dict[t] = 1
len(nng_dict.keys())
```

```python
def find_nnp_tokens(tokens):
    nnp_tokens = []
    for token in tokens:
        if token.tag == "NNP":
            nnp_tokens.append((token.form, token.start, token.end))
    return nnp_tokens

nnp_info = kiwi_tokens.apply(find_nnp_tokens)
nnp_info
```

```python
def critical_omission(ch_srs, df_local):
    result = df_local.copy(deep=True)
    for i in ch_srs.index:
        del_word, start, end = ch_srs.loc[i]
        cur_hyp = result.loc[i, "curHyp"]
        forward = cur_hyp[:start]
        backward = cur_hyp[start:]
        split_back = backward.split(" ")
        del_unit = split_back[0]
        back_mod = " ".join(split_back[1:])
        try:
            if cur_hyp[end] == '"':
                result.loc[i, "curHyp"] = forward.rstrip('"')+back_mod
                result.loc[i, "errors"].append({"severity":"major", "start_idx":start-1, "text":'"'+del_unit, "type":"omission"})
            elif cur_hyp[end] == '」':
                result.loc[i, "curHyp"] = forward.rstrip('「')+back_mod
                result.loc[i, "errors"].append({"severity":"major", "start_idx":start-1, "text":'「'+del_unit, "type":"omission"})
            else:
                result.loc[i, "curHyp"] = (forward+back_mod).strip(" ")
                result.loc[i, "errors"].append({"severity":"major", "start_idx":start, "text":del_unit, "type":"omission"})
        except:
            pass

    return result

co_case = critical_omission(ch3.drop(index=drop_idx, axis=0), temp.drop(index=drop_idx, axis=0))
```

```python
# df = load_data()
def omission_law_name(df_local):
    result = df_local.copy(deep=True)
    for i in result.index:
        cur_hyp = result.loc[i, "curHyp"]
        start = cur_hyp.find("「")
        end = cur_hyp.find("」")
        addition_str = ""
        try:
            while cur_hyp[end+1]!=" ":
                addition_str += cur_hyp[end+1]
                end+=1
        except:
            pass

        result.loc[i, "curHyp"] = cur_hyp[:start].rstrip()+" "+cur_hyp[end+1:].lstrip()
        result.loc[i, "errors"].append({"severity":"critical", "start_idx":start, "text":cur_hyp[start:end+1].strip(), "type":"omission"})
    return result

# oln_df = omission_law_name(df.loc[(df.errors.apply(len) == 0)&(df.KOR.str.contains("「"))])
```
