```python
import pandas as pd
import json
import os
from tqdm.notebook import tqdm

if os.path.exists(r"G:\내 드라이브\KB인턴관련"):
    gdrive_path = r"G:\내 드라이브\KB인턴관련"
else:
    gdrive_path = "/content/drive/MyDrive/KB인턴관련"
```

```python
df = pd.read_csv(os.path.join(gdrive_path, "preprocessing/final.csv"), encoding="euc-kr", index_col=0)
df.head()
```

```python
from kiwipiepy import Kiwi

kiwi = Kiwi(num_workers=0)
kiwi_tokens = df.KOR.apply(lambda x: kiwi.tokenize(x))
kiwi_tokens
```

```python
kiwi_tokens.iloc[0]
```

```python
def count_tag(tokens):
    token_dict = dict()
    for token in tokens:
        try:
            token_dict[token.tag].append(token.form)
        except:
            token_dict[token.tag] = [token.form]
    return token_dict
```

```python
count_tokens = kiwi_tokens.apply(count_tag)
count_tokens
```

```python
nng_dict = dict()
for tokens in count_tokens.apply(lambda x: x["NNG"]):
    for t in tokens:
        try:
            nng_dict[t] += 1
        except:
            nng_dict[t] = 1
len(nng_dict.keys())
```
